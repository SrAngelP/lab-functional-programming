```python
# Define your string handling functions below
# Minimal 3 functions

import os
import re
#import codecs


def cleanhtml(raw_html):
    '''
    This is a function to clean the special html characters in an  given string, returning a clean string.
    
    Arguments:
    
    raw_html = The string you want to clean
    
    Return:
    
    cleantext = A string with the text clean
     
    '''
    
    
    CLEANR = re.compile(r'<.*?>|&([a-z0-9]+|#[0-9]{1,6}|#x[0-9a-f]{1,6});') 
    cleantext = re.sub(CLEANR, '', raw_html)
    return cleantext;

def cleanwords (text):
    '''
    This is s function to clean and separate the words of a string and give you a list of words.
    
    Argument:
    
    text = The text you want to clean and divide.
    
    Return:
    
    cleantext = A list of the words lowered.
    
    '''
    
    
    
    #cleantext =[x.lower() for x in (' '.join((' '.join(' '.join((' '.join((text.rstrip('.').split(':'))).split(';'))).split('\n'))).split(','))).split(' ') if len(x)>0];
    cleantext = re.findall('[A-Za-z]+', text.lower() )
    return cleantext;

def cleantexts(docs):
    '''
    This function clean a list of strings, it returns
    
    Arguments:
    
    docs = A list of strings.
    
    Return:
    
    w_o_t = A list of list with the words of the string.
    
    
    '''
    
    
    w_o_t = [];
    
    for element in docs:
        w_o_t.append(cleanwords(cleanhtml(element)));
        
        
    return w_o_t;

#w_o_t = cleantexts(docs);
#w_o_t

```
